```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import argparse
import ConfigSpace
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_6 = pd.read_csv('config_performances_dataset-6.csv')
df_11= pd.read_csv('config_performances_dataset-11.csv')
df_1457= pd.read_csv('config_performances_dataset-1457.csv')
```

```python
def plot_histograms(dfs, names):
    
    fig, axs = plt.subplots(1,3, figsize=(15,4))
    # axs.size(10)
    # i = 0
    for i, (name, df) in enumerate(zip(names, dfs)):
        print(i, name)
        axs[i].hist(df["score"], bins=50, edgecolor='black')
        axs[i].set_title(name)
        
names = ["dataset-6", "dataset-11", "dataset-1457"]
dfs = [df_6, df_11, df_1457]

plot_histograms(dfs, names)
```

```python
def get_info(df):
    rows, _ = df.shape
    min_score = df['score'].min()
    max_score = df['score'].max()
    print( rows, min_score, max_score)
```

```python
for df in dfs:
    get_info(df)
```
